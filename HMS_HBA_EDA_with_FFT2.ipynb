```python
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science'])

```

```python
eeg = pd.read_parquet('./eeg1.parquet')
orig_cols = eeg.columns
eeg.head()
```

## Look at FFT's of Electrode Sites



```python
# plot the real fft of each column in diff_cols in a 6 x 3 grid
grid_width = 3
grid_height = int(np.ceil(len(orig_cols)/grid_width))
plt.figure(figsize=(10, 20))
for i, col in enumerate(orig_cols):
    plt.subplot(grid_height, grid_width, i+1)
    fft = np.fft.rfft(eeg[col])
    plt.plot(np.abs(fft[1:]))
    plt.title(f'FFT {col}')
plt.tight_layout()
plt.show()
```

### FFT notes on this data
* The sampling rate in time domain is 200 samples/sec (200 Hz)
* The length of this first eeg is 12000 samples, so this represents 60 seconds of data, or 1 minute
* Then each frequency bin is 200/12000 = 1/60 Hz = 0.0167 Hz, and the 1000th bin is 16.7 Hz
* Similarly, 60th bin is 1 hz, 120th bin is 2 hz, etc.

* check the EKG to see if the heart rate makes sense with this sampling rate
*   * 60 beats/min = 1 beat/sec = 1 Hz, so would expect a spike somewhere around bin 60

```python
ekg = eeg.EKG
plt.plot(ekg)
```

```python
# take the real fft of each column in diff_cols
ekg_fft = np.fft.rfft(ekg)
# plot the real fft after removing the dc component
plt.plot(np.abs(ekg_fft[1:]))
plt.title('FFT of EKG')
plt.show()
```

```python
# find bin corresponding to 1hz
bin = np.round(1/.0167)
print(bin)
```

```python
# look at only 1 - 100 bins
plt.plot(np.abs(ekg_fft[1:100]))
plt.title('FFT of EKG')
plt.show()
```

* We do see a spike around bins 50 - 60, which would indicate a heart rate of 50 - 60 beats/min
